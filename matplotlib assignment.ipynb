Ans 2: A scatter plot is a type of data visualization that represents individual data points as dots on a 2D plane. It is commonly used to identify relationships, patterns, or correlations between two numerical variables. Each dot on the scatter plot represents one data point, with the x-coordinate corresponding to one variable and the y-coordinate corresponding to another variable.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)

plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


plt.show()
```

Ans 3:The subplot() function in Matplotlib is used to create multiple plots within a single figure, arranged in a grid-like structure

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line 4")

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Display the plots
plt.show()
```

Ans4:A bar plot is a type of data visualization that represents categorical data using rectangular bars, where the length or height of each bar corresponds to the magnitude of the data.
Bar plots are useful in the following scenarios:

Comparing Categories: Bar plots allow easy visual comparison between different categories or groups.

Showing Distribution: They can display the frequency or distribution of categorical data.

Identifying Trends: Bar plots can help identify trends or patterns in the data.

Highlighting Discrepancies: They can highlight discrepancies or variations among categories.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.title("Company Profits - Bar Plot")
plt.show()
```

```python
# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.title("Company Profits - Horizontal Bar Plot")
plt.show()
```

Ans5:A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of numerical data through quartiles. It displays a summary of the data's central tendency, spread, and identifies potential outliers
Box plots are used for the following purposes:

Visualizing Distribution: They provide a concise way to understand the distribution of data, particularly the spread and skewness.

Comparing Distributions: Box plots allow for easy comparison between multiple datasets.

Identifying Outliers: They help in identifying potential outliers in the data.

Showing Central Tendency: Box plots indicate the median, which is the central tendency of the data.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel("Box")
plt.ylabel("Values")
plt.title("Box Plot Example")
plt.show()
```
