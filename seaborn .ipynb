Ans1: Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics.

Here are five plots that can be created using Seaborn along with their uses:

a:Scatter Plot

Use: Scatter plots are used to visualize the relationship between two continuous variables. They show individual data points as dots on a 2D plane, helping to identify patterns and correlations between the variables.

b:Bar Plot

Use: Bar plots are useful for comparing categorical data or showing the distribution of a single categorical variable. They display rectangular bars, where the height of each bar represents the magnitude of the data.

c:Histogram

Use: Histograms are used to visualize the distribution of a single continuous variable. They group data into bins and display the frequency of data points falling into each bin as bars.

d:Heatmap

Use: Heatmaps are used to represent the intensity of values within a 2D matrix. They use colors to visualize the magnitude of data and are commonly used in correlation matrices or to display patterns in large datasets.

e:Box Plot

Use: Box plots (box-and-whisker plots) show the distribution of numerical data through quartiles. They are used to compare distributions, identify outliers, and understand the spread and central tendency of data

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_dataset = sns.load_dataset("fmri")

# Plot the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_dataset)

# Add title and labels
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_dataset = sns.load_dataset("titanic")

# Plot the box plots
plt.figure(figsize=(10, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_dataset)
plt.title("Box Plot of Age by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Age")

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_dataset)
plt.title("Box Plot of Fare by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_dataset = sns.load_dataset("diamonds")

# Plot the histogram with hue for 'cut'
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_dataset, x='price', hue='cut', kde=True, bins=30)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_dataset = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
flights_matrix = flights_dataset.pivot("month", "year", "passengers")

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_matrix, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Flights Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```
